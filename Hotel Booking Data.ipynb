```python
# outliers
mydata.boxplot(column=['lead_time'])
```

```python
def remove_outlier(col):                    #code for calculating and removing outliers
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range=Q1-1.5*IQR
    upper_range=Q3+1.5*IQR
    return lower_range,upper_range
```

```python
low_leadtime,high_leadtime=remove_outlier(mydata['lead_time'])                        # removing leadtime from data
mydata['lead_time']=np.where(mydata['lead_time']>high_leadtime,high_leadtime,mydata['lead_time'])
mydata['lead_time']=np.where(mydata['lead_time']<low_leadtime,low_leadtime,mydata['lead_time'])
```
